```python
def draw_bounding_boxes(image_path, xml_path):
    # Load image
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert color space for matplotlib

    # Parse XML
    tree = ET.parse(xml_path)
    root = tree.getroot()

    # Iterate over each 'object' element
    for obj in root.iter('object'):
        
        # get object name
        obj_name = obj.find('name').text
        
        # Get bounding box coordinates
        bndbox = obj.find('bndbox')
        xmin = int(bndbox.find('xmin').text)
        ymin = int(bndbox.find('ymin').text)
        xmax = int(bndbox.find('xmax').text)
        ymax = int(bndbox.find('ymax').text)

        # Draw bounding box
        img = cv2.rectangle(img, (xmin, ymin), (xmax, ymax), (0, 255, 0), 2)

        # Put object name
        cv2.putText(img, obj_name, (xmin, ymin-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (36,255,12), 2)

    plt.figure(figsize=(12,12))
    # Display image
    plt.imshow(img)
    plt.show()
```

```python
# Call the function
draw_bounding_boxes('JPEGImages/frames_38 (237).jpg', 'Annotations/frames_38 (237).xml')
```
